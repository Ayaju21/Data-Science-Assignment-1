### Import Libraries and Load Kaggle Dataset 

```python
# Import Pandas and NumPy libraries
import pandas as pd
import numpy as np

# Load the dataset from Kaggle into a Pandas DataFrame
df = pd.read_csv('train.csv')

# Select 10 columns from the dataset for analysis
cols = ['SalePrice', 'LotArea', 'OverallQual', 'YearBuilt', 'TotalBsmtSF',
        '1stFlrSF', 'GrLivArea', 'FullBath', 'BedroomAbvGr', 'GarageCars']

# Display the first 5 rows of the selected columns
df[cols].head()
```

## Summary Statistics of Selected Columns 📊 

- **Mean :**
  
  The average of the values in the SalePrice "Price of the house" column.

```python
means = df['SalePrice'].mean()
print("Mean (SalePrice):", means)
```

 This is a single representative value that summarizes all the house prices in the dataset. It shows the typical price of a house in the data. By comparing individual house prices to this mean, we can see which houses are more expensive or cheaper than the typical value, which helps in understanding the overall price distribution.

- **Standard Deviation:**
  
     Measures how spread out the values are in the LotArea "Size of the land" column.

  

```python
std = df['LotArea'].std()
print("Standard Deviation (LotArea):", std)
```

This shows that the sizes of the house lots vary widely in the dataset. Some houses are on much larger lots while others are on smaller ones. A high standard deviation like this indicates that there is a lot of diversity in the land sizes among the houses.

- **Count :**
  
   The number of non-missing values in the BedroomAbvGr "Number of bedrooms above ground" column.  

```python
count = df['BedroomAbvGr'].count()
print("Count (BedroomAbvGr):", count)
```

This count shows that every house has data for the number of bedrooms above ground column, so there are no missing values.

- **Median :**
  
  The middle value when the data in the GrLivArea "Above-ground living area" column is sorted.

```python
median = df['GrLivArea'].median()
print("Median (GrLivArea):", median)
```

This means that half of the houses in the dataset have an above-ground living area smaller than 1464.0 square feet, and the other half have a larger living area.  It represents a typical house size in terms of living space above ground and helps us understand the distribution of house sizes in the dataset.

- **75th Percentile :**
  
   The value below which 75% of the TotalBsmtSF "Total Basement Square Feet" data falls.


```python
percentile_75 = df['TotalBsmtSF'].quantile(0.75)
print("75th Percentile (TotalBsmtSF):", percentile_75)
```

This means that 75% of the houses in the dataset have a basement area of 1298.25 square feet or less, and the remaining 25% have a larger basement. It helps us understand the distribution of basement sizes, showing which houses have among the largest 25% of basement space in the dataset.

- **Question 3 :** Do houses with higher quality scores sell for higher prices ?

```python
quality_price = df.groupby('OverallQual')['SalePrice'].mean()
print(quality_price)
```

- **Question 4 :** What are the smallest and largest lot sizes ?

```python
max_lot = df['LotArea'].max()
min_lot = df['LotArea'].min()
print("Largest Lot Area:", max_lot)
print("Smallest Lot Area:", min_lot)
```

The largest lot is the biggest property, and the smallest lot is the tiniest. This helps buyers and sellers understand the range of land sizes available.

- **Question 5 :** How does the number of bedrooms affect house price ?
